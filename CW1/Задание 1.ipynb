```python
def reading():
    sym = "0123456789.,?!…:;()[]-—_|/\"'«»*{}<>@#$%^&№“”"
    e = []
    m = []
    with open('esenin.txt', encoding='utf-8') as f:
        for line in f:
            words_e = line.strip().lower().split()
            for word in words_e:
                e.append(word.strip(sym))
        for i in e:
            if len(i) < 1:
                e.remove(i)
        esenin = set(e)
    with open('mayak.txt', encoding='utf-8') as g:
        for line in g:
            words_m = line.strip().lower().split()
            for word in words_m:
                m.append(word.strip(sym))
        for i in m:
                if len(i) < 1:
                    m.remove(i)
        mayak = set(m)
    a, b = esenin, mayak
    return a, b
```

```python
a, b = reading()
def zadanie(a, b):
    print("Общие словоформы:", a & b)
    print("Уникальные словоформы у Есенина:", a ^ b, "\nУникальные словоформы у Маяковского:", b ^ a)
    
zadanie(a, b)
```
