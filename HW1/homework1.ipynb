```python
import random

def choice():
    ## выбор категории слов
    vvod = input('Выберите одну из категорий, набрав 1, 2 или 3 соответственно:\
    \n1) Растения \n2) Писатели и поэты \n3) Животные\n')
    if  vvod == '1':
        with open ('1.txt', 'r', encoding = 'utf-8') as f:
            list = f.readlines()
            i = random.choice(list)
    elif vvod == '2':
        with open ('2.txt', 'r', encoding = 'utf-8') as g:
            list = g.readlines()
            i = random.choice(list)
    elif vvod == '3':
        with open ('3.txt', 'r', encoding = 'utf-8') as h:
            list = h.readlines()
            i = random.choice(list)
    ## если не 1, 2 или 3, то игра начинается заново
    else:
        print('Будьте внимательны, попробуйте снова.')
        game(choice())
    ## создаётся "скелет" слова
    word = []
    for l in i.strip().lower():
        word.append(l)
    ## word - это загаданное слово по буквам 
    return word
```

```python
## функция берёт список из букв и проводит дальнейшие операции с ним

def game(word):
    n = len(word)
    slovo = []
    p = 6               ## количество попыток
    for number in range(n):
        slovo.append('_')
    ## slovo - это список из "_", количество которых равно колличеству букв в слове
    whole = ''.join(word)
    ## whole - это строка = загаданное слово
    print('У вас есть ', p, ' попыток, чтобы угадать слово из ', n, ' букв:\n', ' '.join(slovo), sep='')
    alph = 'абвгдеёжзийклмнопрстуфхцчшщьыъэюя'
    x = 1               ## счётчик попыток
    tries = []          ## список самих попыток
    while x <= 6:
        if '_' not in slovo:           ## проверка на то, отгаданы ли все буквы
            print('Вы победили!\n')
            if input('Хотите начать заново? Нажмите Enter.') == '':
                game(choice())
        letter = input('\nВведите букву: ')
        if letter.lower() in tries:
            print('Вы уже вводили эту букву. Попробуйте другую.')
        elif letter in alph:
            tries.append(letter)
            if letter in word:
                for number in range(len(whole)):
                    if word[int(number)] == letter:
                        slovo.pop(int(number))              ## удаляет '_' с нужным индексом
                        slovo.insert(int(number), letter)   ## добавляет букву на место удалённого знака '_'
                print('Буква есть в слове:', ' '.join(slovo))
                continue
            else:
                x += 1
                p -= 1
                if p == 5:
                    print('Такой буквы нет в слове. У вас осталось 5 попыток.')
                    print(' __\n|  |\n|  0\n|\n|')
                elif p == 4:
                    print('Такой буквы нет в слове. У вас осталось 4 попытки.')
                    print(' __\n|  |\n| \\0\n|\n|')
                elif p == 3:
                    print('Такой буквы нет в слове. У вас осталось 3 попытки.')
                    print(' __\n|  |\n| \\0/\n|\n|')
                elif p == 2:
                    print('Такой буквы нет в слове. У вас осталось 2 попытки.')
                    print(' __\n|  |\n| \\0/\n|  |\n|')
                elif p == 1:
                    print('Такой буквы нет в слове. У вас осталось 1 попытка.')
                    print(' __\n|  |\n| \\0/\n|  |\n| /')
                elif p == 0:
                    print('Такой буквы нет в слове. У вас осталось 0 попыток.')
                    print(' __\n|  |\n| \\0/\n|  |\n| / \\')
                continue
        else:
            print('Это не буква или буква не из русского алфавита, будьте внимательнее.')
            continue
    else:
        print('Попытки закончились. Вы проиграли. Это было слово:', whole, '\n')
        if input('Хотите начать заново? Нажмите Enter.') == '':
            game(choice())
```

```python
def main():
    game(choice())
    
if __name__=='__main__':
    main()
```
